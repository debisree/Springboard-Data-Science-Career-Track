```python
#The problem - Empire State Building step gambling!

#import
import numpy as np
import matplotlib.pyplot as plt

#seed
np.random.seed(123)

#Simulate the random walk 500 times

all_walks = []
for i in range(500):

#initiate the random walk

    random_walk = [0]
    for x in range(100):
        step = random_walk[-1]

#Roll the dice
        dice = np.random.randint(1,7)

#The step counts : Given in the problem statement
        
        if dice <= 2:
            step = max(0,step-1)
        elif dice <=5:
            step = step+1
            else:
                step = step+np.random.randint(1,7)
                
# Implement the clumsiness

        if np.random.rand() <= 0.001:
            step = 0
            
        random_walk.append(step)
        all_walks.append(random_walk)
        
#convert all_walks into NumPy array    

    np_aw = np.array(all_walks)
    
#Transpose the array

    np_aw_t= np.transpose(np_aw)
    
#To see the end points of all walks

     ends = np_aw_t[-1,:]

#Plot the histogram

      plt.hist(ends)
      plt.show()
    
 #The solution of the original question "what are the odds that you will reach 60 steps high?"
     print(np.mean(ends >= 60))
     
```
