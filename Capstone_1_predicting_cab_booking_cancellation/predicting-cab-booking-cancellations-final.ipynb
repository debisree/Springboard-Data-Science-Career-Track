## Introduction:

### 1. The Problem statement:

The business problem addressed here is to improve the customer service for Bangalore metropolitan (India) based cab company called **YourCabs**. The problem is that, a certain percentage of booking gets canceled by the company due to the unavailability of a car, and the cancellations occur at a time when the trip is about to start. Therefore it causes passengers inconvenience and a bad reputation for the company. So, the challenge is to build a predictive model, which would classify the upcoming bookings as, if they would eventually get cancelled due to car unavailability, or not. So this is a classification problem.

### 8. Applying Machine Learning models and comparing their performances:

This is a classification problem, in supervised learning. Here we have used the following classification models:
* Logistic Regression 
* K-Nearest Neighbor (KNN)
* Support vector machine (SVM)
* Random Forest 
* Naive Bayes
* Gradient Boost

Evaluating the performance of a model by training and testing on the same dataset can lead to the overfitting. Hence the model evaluation is based on splitting the dataset into train and validation set. However, the performance of the prediction result depends upon the random choice of the pair of (train, validation) set. To overcome, the Cross-Validation procedure is used where under the k-fold CV approach, the training set is split into k smaller sets, where a model is trained using k-1 of the folds as training data, and the model is validated on the remaining part.

**Classification/Confusion Matrix:**  This matrix summarizes the correct and incorrect classifications that a classifier produced for a certain dataset. Rows and columns of the classification matrix correspond to the true and predicted classes respectively. The two diagonal cells (upper left, lower right) give the number of correct classifications, where the predicted class coincides with the actual class of the observation. The off diagonal cells gives the count of the misclassification. The classification matrix gives estimates of the true classification and misclassification rates.


We applied different ML models above and evaluated their performances in terms of ROC-AUC score for both the training and test data. Here we have tabulated the scores and plotted them.

Clearly, the **Gradient Boost**, and the **Random Forest** are the two best performing models. 


Both the models are similar in nature, based on decision trees. Both are ensemble learning methods and predict (regression or classification) by combining the outputs from individual trees. The two main differences are:

* How trees are built: random forests builds each tree independently while gradient boosting builds one tree at a time. This additive model (ensemble) works in a forward stage-wise manner, introducing a weak learner to improve the shortcomings of existing weak learners. 
* Combining results: random forests combine results at the end of the process (by averaging or "majority rules") while gradient boosting combines results along the way.

### 10. Conclusion:

In the two final cancellation prediction (result) files **'final_result_gb.csv'**, and **'final_result_rf.csv'**, there are two columns, named **User ID** and **Car_cancellation**. It shows the prediction for cab booking cancellations (0 for no cancellation; 1 for cancellation) corresponding to the user IDs. 

This project aims to predict the cancellations, to prevent/reduce them, which would eventually improve the business of the cab company. It has been stated that most of the times, the cancellations are due to the unavailability of cars. Which means the company fails to meet the right demand-supply ratio. From the given data, it is difficult to pinpoint, wether a cancellation is due to the car unavailability (from the company side) or from the customer. Following findings are the conclusions, in a nutshell, found from the analysis:

* **Weekends:** We see that in general in the weekends, the cab booking rates are higher than the weekdays, which makes sense. However, in terms of cancellations, we can see that the maximum (18.5%) bookings are canceled on Fridays, followed by Sundays (18.4%).  So, to meet the demand-supply ratio, the most generic prescription is, to increase the number of cabs during the weekends. 

* **Weekdays:**  We see that the most of the times, cab demands are high in two typical times of the days. One is in the morning; another is in the evening. These are the busy/rush/ office hours on a weekday. The cancellation pattern also follows the same. So, the cancellations are due to unavailability of cars in these hours. The company cannot meet the correct demand-supply ratio. So again, the prescription is to increase the number of vehicles in these two-time slots roughly by 10-20%.

* **Monthly business pattern:** The data trend shows, that the cab demand is maximum in the month of July-October, followed by May-June. Also, most of the cancellations are in October-Nov. This could be for many reasons:
  * The typical monsoon season in Bangalore lasts in these months. So, people can book more cabs during the monsoon days. Also, because of the weather, it is a fair chance for them to cancel the bookings. For reference, one can see this: http://bengaluru.citizenmatters.in/some-lessons-for-bengaluru-from-100-years-of-rain-data-8013
  * Most of the religious festivals and holidays are during August-October. People go out more these days and book more cabs. Which could result in more cancellations too. People can cancel their plans at the end moment. Heavy traffic congestion due to festival season, can lead to more cab cancellations. For reference: https://www.karnataka.com/govt/list-of-2019-government-holidays/
  

* **Area/Location:** We have seen that there are specific area/GPS locations and routes, for which the cancellation rates are high. The reason could be the traffic scenario. Bad traffic congestion can give rise to late arrival or cancellations. The cab company must keep on eyes on the live traffic update of the city, and increase the number of cabs in those areas.

* Lastly, to prevent or check cancellations from the customers, the company must take the following prescription:
  * Run the model at every one-hour interval
  * Call the customer who is flagged by the model
  * Confirm with the customer if the booking will be canceled or not
  * Send cab only after the confirmation from the customer
